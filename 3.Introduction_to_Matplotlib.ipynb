### Importing the necessary libraries

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

```%matplotlib inline``` is the command for static plot inside jupyter notebook. Depending upon the version of jupyter notebook one may run this command. For latest version of jupyter notebook this is not required. 

For interactive plots one can use the magic command ```%matplotlib notebook```.

### Creating data for plot

````python
x = np.linspace(0,4*np.pi,1000)
x.shape
````

````python
y = np.sin(x)
y.shape
print(x)
````

### Line Plot

````python
plt.figure(figsize = (10,8))
plt.plot(x,y,'b')
plt.grid()
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.title('sin(X) vs X');
````

### Simultaneously plotting multiple lineplots in the same figure

````python
y1 = np.sin(x)
y2 = np.cos(x)
````

````python
plt.figure(figsize = (10,6))
plt.plot(x,y1,c = 'b',label = 'sin')
plt.plot(x,y2,c = 'r',label = 'cos')
plt.grid()
plt.legend()
plt.xlabel('X')
plt.ylabel('sin or cos x')
plt.title('My_plots')
````

#### Few more examples with line plot

````python
y1 = 2*y1
y2 = 3*y2
plt.figure(figsize = (8,5))
plt.plot(y1,y2)
````

### Subplots

````python
x = np.linspace(1,10,1000)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x)
y4 = np.exp(x)

plt.figure(figsize = (8,8))
plt.suptitle('My plots')

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Sin(X)")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Cos(x)')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('log(x)')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('exp(x)');
````

### Scatter Plot

````python
x = np.random.rand(30)
````

````python
y = np.random.rand(30)
````

````python
plt.scatter(x,y)
````

### Bar Plot 

````python
values = np.asarray([10,15,5,19,23,17])

plt.bar(x = np.arange(1,7), height = values)
````

````python
plt.figure(figsize = (8,6))

plt.bar(x = ['d1','d2','d3','d4','d5','d6'], height = values, color = 'r');
````

### Horizontal Bar Plot

````python
plt.barh(y = np.arange(1,7), width = values)
````

### Histogram Plot

````python
n = np.random.randn(1000000)

plt.figure(figsize = (10,5))

plt.hist(n, bins = 1000, edgecolor = 'r');
````

### 3D Scatter Plot

````python
from mpl_toolkits.mplot3d import Axes3D
````

````python
z = np.random.rand(30)
````

````python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter3D(x,y,z)
plt.show()
````

### 3D surface plot

````python
(x1,y1) = np.meshgrid(x,y)
````

````python
z1 = 2*x1 + 3*y1
````

````python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x1,y1,z1)
plt.show()
````
